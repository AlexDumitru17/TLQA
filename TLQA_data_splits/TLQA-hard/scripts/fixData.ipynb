```python
import re

def extract_years(question):
    # Find all 4-digit numbers in the question
    years = re.findall(r'\b\d{4}\b', question)
    
    if len(years) >= 2:
        return pd.Series({'startYear': int(years[0]), 'endYear': int(years[1])})
    elif len(years) == 1:
        # If only one year is found, use it for both start and end
        return pd.Series({'startYear': int(years[0]), 'endYear': int(years[0])})
    else:
        return pd.Series({'startYear': None, 'endYear': None})

# Apply the function to create new columns
time_intervals_df[['startYear', 'endYear']] = time_intervals_df['newQuestion'].apply(extract_years)
```

```python
def is_within_interval(year_str, start_year, end_year):
    # Convert to string if it's an integer
    year_str = str(year_str).strip()
    
    # Handle empty strings
    if not year_str:
        return False, None
    
    # Handle single year
    if year_str.isdigit():
        year = int(year_str)
        if start_year <= year <= end_year:
            return True, str(year)
        return False, None
    
    # Handle year intervals
    years = year_str.split('-')
    if len(years) == 2 and years[0].strip() and years[1].strip():
        try:
            interval_start = int(years[0])
            interval_end = int(years[1])
            
            # Check if there's any overlap
            if not (interval_end < start_year or interval_start > end_year):
                # Adjust the interval to fit within start_year and end_year
                adjusted_start = max(interval_start, start_year)
                adjusted_end = min(interval_end, end_year)
                
                # If start and end years are the same, return just the year
                if adjusted_start == adjusted_end:
                    return True, str(adjusted_start)
                return True, f"{adjusted_start}-{adjusted_end}"
            return False, None
        except ValueError:
            return False, None
    
    return False, None

def filter_answers_by_timeframe(answers_dict, start_year, end_year):
    filtered_answers = {}
    
    for team, years in answers_dict.items():
        # Convert to string if it's an integer
        years = str(years)
        
        # Handle multiple intervals separated by comma
        year_spans = years.split(', ')
        valid_spans = []
        
        for span in year_spans:
            is_valid, adjusted_span = is_within_interval(span, start_year, end_year)
            if is_valid:
                valid_spans.append(adjusted_span)
        
        if valid_spans:
            filtered_answers[team] = ', '.join(valid_spans)
    
    return filtered_answers

# Apply the filter to each row
time_intervals_df['fixedAnswers'] = time_intervals_df.apply(
    lambda row: filter_answers_by_timeframe(
        row['answers'], 
        row['startYear'], 
        row['endYear']
    ), 
    axis=1
)
```
